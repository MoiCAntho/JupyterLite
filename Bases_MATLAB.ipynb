# Bases de MATLAB

## Opération élémentaires
Les opérations classiques sont définies comme il suit :
- Somme : +
- Différence : -
- Multiplication : *
- Quotient : /
- Puissance : ^

Notons que l'utilisation des parenthèse est analogue à l'utilisation courante pour les priorités de calcul.

```octave
(3*2)/(5+3)
```
